```
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn]))) 
      
import pandas 
from sklearn import linear_model
import statsmodels.api as sm
import matplotlib.pyplot as plt
df =pandas.read_csv('cars.csv')
#Visualize linear relationship between Weight and CO2
plt.scatter(df['Weight'], df['CO2'], color='red')
plt.title('CO2 Vs Weight', fontsize=14)
plt.xlabel('Weight', fontsize=14)
plt.ylabel('CO2', fontsize=14)
plt.grid(True)
plt.show()


#Visualize linear relationship between Unemployment_Rate and Stock_Index_Price
plt.scatter(df['Volume'], df['CO2'], color='green')
plt.title('CO2 Vs Volume', fontsize=14)
plt.xlabel('Volume', fontsize=14)
plt.ylabel('CO2', fontsize=14)
plt.grid(True)
plt.show()

X = df[['Weight','Volume']] # here we have 2 variables for multiple regression. Ifyou just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] forexample.Alternatively, you may add additional variables within the brackets
Y = df['CO2']


# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X,Y)
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)



# prediction with sklearn
 
print ('Predicted CO2: \n',regr.predict([[3300, 1300]]))
```
